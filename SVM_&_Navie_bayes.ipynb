1. What is Information Gain, and how is it used in Decision Trees?

Information Gain

Information Gain is a measure used in decision tree algorithms to determine the best attribute for splitting a dataset. It is based on the concept of entropy, which measures the level of uncertainty or impurity in the data. If all instances in a dataset belong to a single class, the entropy is zero, indicating perfect purity. However, if the data is evenly distributed among different classes, the entropy is high, showing greater randomness. Information Gain calculates how much this uncertainty is reduced after splitting the dataset based on a particular feature. It is computed by subtracting the weighted entropy of the subsets formed by a split from the original entropy of the dataset.

In decision tree construction, especially in algorithms like ID3, Information Gain is used as a criterion for selecting the splitting attribute at each node. The algorithm calculates the Information Gain for all available features and chooses the one with the highest value, as it provides the maximum reduction in entropy. This process is repeated recursively for each subset until the data is perfectly classified or no further meaningful splits can be made. Thus, Information Gain plays a crucial role in building efficient and accurate decision trees by ensuring that each split improves the classification performance.


2. What is the difference between Gini Impurity and Entropy?

Gini Impurity and Entropy are two commonly used measures to evaluate the quality of splits in decision tree algorithms. Both are used to measure the impurity or randomness in a dataset, but they differ in their calculation, interpretation, and practical usage. Entropy is based on information theory and measures impurity using a logarithmic function. Its value ranges from 0 (pure node) to 1 (for binary classification with maximum impurity). It is used in algorithms like ID3 and C4.5. Entropy tends to penalize more heavily when class probabilities are uneven and is slightly more sensitive to changes in class distribution.

Gini Impurity, on the other hand, measures the probability of incorrectly classifying a randomly chosen element if it were labeled according to the class distribution of the node. Its value also ranges from 0 (pure node) to a maximum of 0.5 in binary classification. Gini Impurity is used in algorithms like CART. Unlike Entropy, Gini does not use logarithms, which makes it computationally faster and simpler to calculate. In practice, both measures often produce very similar trees, but Gini is generally preferred when computational efficiency is important, while Entropy is sometimes chosen when a more theoretically grounded measure from information theory is desired. Overall, the difference lies mainly in their mathematical formulation, computational complexity, and slight variation in sensitivity to class distribution.


3. What is Pre-Pruning in Decision Trees?

Pre-pruning in decision trees is a technique used to stop the growth of a tree early, before it becomes too complex and overfits the training data. Instead of allowing the tree to grow fully and then trimming it later, pre-pruning applies certain stopping conditions during the tree-building process. This helps prevent the model from learning noise and minor fluctuations in the dataset. The main goal of pre-pruning is to improve the model’s ability to generalize to unseen data.

In pre-pruning, the algorithm checks specific criteria before making a split at a node. If the split does not meet the predefined conditions, the node is declared a leaf node and no further splitting occurs. Common stopping criteria include setting a maximum depth for the tree, requiring a minimum number of samples in a node to allow splitting, specifying a minimum information gain or reduction in impurity, or limiting the minimum number of samples required in a leaf node. These constraints control the growth of the tree and reduce its complexity.

Pre-pruning is commonly used in decision tree algorithms such as CART. The main advantage of pre-pruning is that it reduces training time and prevents overfitting early. However, a disadvantage is that it may stop the tree too soon, leading to underfitting if important splits are prevented. Therefore, the stopping criteria must be carefully chosen to maintain a balance between model simplicity and predictive accuracy.


4. Write a Python program to train a Decision Tree Classifier using Gini Impurity as the criterion and print the feature importances (practical).  

```python
# Import required libraries
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Load dataset (Iris dataset)
data = load_iris()
X = data.data
y = data.target
feature_names = data.feature_names

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Create Decision Tree Classifier using Gini Impurity
clf = DecisionTreeClassifier(criterion='gini', random_state=42)

# Train the model
clf.fit(X_train, y_train)

# Print feature importances
print("Feature Importances:")
for name, importance in zip(feature_names, clf.feature_importances_):
    print(f"{name}: {importance:.4f}")
```

5. What is a Support Vector Machine (SVM)?

A **Support Vector Machine (SVM)** is a supervised machine learning algorithm used for classification and regression tasks. It is mainly used for classification problems and works by finding the optimal boundary (called a **hyperplane**) that separates data points belonging to different classes. The main objective of SVM is to maximize the margin, which is the distance between the hyperplane and the nearest data points from each class. These nearest data points are called **support vectors**, and they play a crucial role in defining the decision boundary.

In a two-dimensional space, the hyperplane is simply a straight line, but in higher dimensions, it becomes a plane or a higher-dimensional surface. SVM tries to choose the hyperplane that not only separates the classes but does so with the maximum possible margin, which improves the model’s generalization ability and reduces the risk of overfitting.

SVM can handle both linearly separable and non-linearly separable data. For non-linear data, it uses a technique called the **kernel trick**, which transforms the data into a higher-dimensional space where it becomes easier to separate. Common kernels include linear, polynomial, and radial basis function (RBF) kernels.

SVM is a powerful and effective algorithm, especially for high-dimensional datasets, because it focuses only on the critical data points (support vectors) and aims to build a robust and well-generalized decision boundary.


6. What is the Kernel Trick in SVM?

The **Kernel Trick** in Support Vector Machines (SVM) is a technique used to handle non-linearly separable data. In many real-world problems, the data cannot be separated by a straight line (or simple hyperplane) in its original feature space. The kernel trick allows SVM to transform the data into a higher-dimensional space where a linear separation becomes possible, without explicitly computing the transformation.

Normally, mapping data to a higher dimension would be computationally expensive. However, the kernel trick avoids this by using a **kernel function**, which directly computes the inner product of data points in the higher-dimensional space. This means the algorithm can work as if the data were transformed, without actually performing the transformation. This makes SVM both powerful and efficient.

Common kernel functions include:

* **Linear Kernel** – Used when data is already linearly separable.
* **Polynomial Kernel** – Introduces curved decision boundaries.
* **Radial Basis Function (RBF) Kernel** – Creates flexible, non-linear boundaries and is widely used.
* **Sigmoid Kernel** – Similar to neural network activation functions.

In simple terms, the kernel trick allows SVM to draw complex decision boundaries in the original space by mathematically projecting the data into a higher dimension, making it possible to solve non-linear classification problems effectively.


7. Write a Python program to train two SVM classifiers with Linear and RBF kernels on the Wine dataset, then compare their accuracies.

Hint:Use SVC(kernel='linear') and SVC(kernel='rbf'), then compare accuracy scores after fitting on the same dataset.

```python
# Import required libraries
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load Wine dataset
data = load_wine()
X = data.data
y = data.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Feature scaling (important for SVM)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Train SVM with Linear Kernel
svm_linear = SVC(kernel='linear', random_state=42)
svm_linear.fit(X_train, y_train)
y_pred_linear = svm_linear.predict(X_test)
accuracy_linear = accuracy_score(y_test, y_pred_linear)

# Train SVM with RBF Kernel
svm_rbf = SVC(kernel='rbf', random_state=42)
svm_rbf.fit(X_train, y_train)
y_pred_rbf = svm_rbf.predict(X_test)
accuracy_rbf = accuracy_score(y_test, y_pred_rbf)

# Print accuracies
print("Accuracy with Linear Kernel:", accuracy_linear)
print("Accuracy with RBF Kernel:", accuracy_rbf)
```

8. What is the Naïve Bayes classifier, and why is it called "Naïve"?

The **Naïve Bayes classifier** is a supervised machine learning algorithm based on **Bayes’ Theorem**, which is used for classification tasks. It calculates the probability of a data point belonging to a particular class based on the prior probability of the class and the likelihood of the observed features. The classifier assigns the data point to the class with the highest posterior probability. Naïve Bayes is widely used in applications such as spam detection, sentiment analysis, and document classification because it is simple, fast, and performs well even with large datasets.

It is called “naïve” because it makes a strong assumption that all features are independent of each other given the class label. In other words, it assumes that the presence or absence of one feature does not affect the presence or absence of another feature. In real-world situations, this assumption is often unrealistic because features are usually related. However, despite this simplifying assumption, the algorithm performs surprisingly well in many practical problems.

In simple terms, Naïve Bayes uses probability to make predictions and is considered “naïve” because it assumes complete independence among features, which simplifies calculations and makes the model computationally efficient.


9. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes

Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes are three different variants of the Naïve Bayes classifier, and they differ mainly in the type of data they are designed to handle and the probability distribution they assume for the features.

**Gaussian Naïve Bayes** assumes that the continuous features follow a normal (Gaussian) distribution. It is mainly used when the input variables are continuous numerical values, such as height, weight, or measurements in a dataset. For each feature, it estimates the mean and standard deviation for each class and then calculates the probability using the Gaussian probability density function. This version is suitable for datasets where features are real-valued and normally distributed.

**Multinomial Naïve Bayes** is used for discrete count data. It is commonly applied in text classification problems such as spam detection or sentiment analysis. In this model, features typically represent word frequencies or term counts in a document. Instead of assuming a normal distribution, it models the probability of counts using a multinomial distribution. It works well when features represent the number of times an event occurs.

**Bernoulli Naïve Bayes** is designed for binary or boolean features. Each feature is assumed to be either present (1) or absent (0). It is also often used in text classification, but instead of using word counts, it considers whether a word appears in a document or not. This makes it suitable when the presence or absence of a feature is more important than how many times it appears.

The key difference lies in the type of data they handle: Gaussian Naïve Bayes is used for continuous data, Multinomial Naïve Bayes is used for count-based discrete data, and Bernoulli Naïve Bayes is used for binary data. Choosing the appropriate variant depends on the nature of the input features in the dataset.


10. Write a Python program to train a Gaussian Naïve Bayes classifier on the Breast Cancer dataset and evaluate accuracy.

```python
# Import required libraries
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# Load Breast Cancer dataset
data = load_breast_cancer()
X = data.data
y = data.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Feature scaling (optional but recommended)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Create Gaussian Naïve Bayes classifier
gnb = GaussianNB()

# Train the model
gnb.fit(X_train, y_train)

# Make predictions
y_pred = gnb.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy of Gaussian Naïve Bayes on Breast Cancer dataset:", accuracy)
```
